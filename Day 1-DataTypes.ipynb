```python
#CLASS-user-defined prototype or blueprint of an object
#class clsname - 2 properties-attribute and behavior
class Student:
    def __init__(self,college):
        print("College: ",college)
    #Methods-Similar to function
    def details(self,name,course):
        #self is used to access the attributes or behavior var value
        print("Name: ",name)
        print("Course: ",course)
    def perform(x,grade):
        print("Rank: ",grade)
obj=Student("RNSIT")
obj.details("Deethya","ISE")
```

```python
#Object-Instance of the class
s=Student()
```

```python
s.details("Preethu","CS")
```

```python
s1=Student()
s1.details("Karthik","IT")
s1.perform("A")
```

```python
#Inheritance-Base class and Derived class
#Derived class can access base cls properties
#base class can't access derived class properties
#single Inheritance-1 base class and 1 derived class
class college:#Base class
    def __init__(self,college):
        print("College: ",college)
    #Methods-Similar to function
class Student(college):#Derived class
    def details(self,name,course):
        #self is used to access the attributes or behavior var value
        print("Name: ",name)
        print("Course: ",course)
    def perform(x,grade):
        print("Rank: ",grade)
obj=Student("RNSIT")
obj.details("Deethya","ISE")
```

```python
#Inheritance-Base class and Derived class
#Derived class can access base cls properties
#base class can't access derived class properties
#Multiple Inheritance-1 base class and Multiple derived class
class college:#Base class
    def __init__(self,college):
        print("College: ",college)
    #Methods-Similar to function
class Student(college):#Derived class
    def details(self,name,course):
        #self is used to access the attributes or behavior var value
        print("Name: ",name)
        print("Course: ",course)
class marks(Student):
    def perform(x,grade):
        print("Rank: ",grade)
obj=Student("RNSIT")
obj.details("Deethya","ISE")
obj1=marks("A")
```

```python
#Multiple Inheritance
class Add:#Base class
    def add(x,a,b):
        print(a+b)
class Sub:#Base class
    def sub(x,a,b):
        print(a-b)
class Mul:#Base class
    def mul(x,a,b):
        print(a*b)
class Div(Add,Sub,Mul):
    def div(x,a,b):
        print(a/b)
```

```python
#Polymorphism-overriding the data or many forms
class Add:#Base class
    def add(x,a,b):
        print(a+b)
class Sub:#Base class
    def sub(x,a,b):
        print(a-b)
class Mul:#Base class
    def mul(x,a,b):
        print(a*b)
class Div(Add,Sub,Mul):#Derived class
    def add(x,a,b):
        print(a/b)
    def add(x):
        print("Add")
```

```python
# 1. Finding greatest of three numbers
x = int(input("Enter value of x:"))
y = int(input("Enter value of y:"))
z = int(input("Enter value of z:"))

if x>=y and x>=z:
    L=x
elif y>=x and y>=z:
    L=y
else:
    L=z

print("The greatest of the three numbers is: "+str(L))
```

```python
# 2. if a<=1000(give a discount of 10%),a<=2000(give a discount of 20%),a<=3000(give a discount of 30%)
a = int(input("Enter the price of the item: "))
if a<=1000:
    a=a-(a*0.10)
    print("The Discounted price of a is ",format(a))
    
elif a<=2000:
    a=a-(a*0.20)
    print("The Discounted price of a is ",format(a))
    
elif a<=3000:
    a=a-(a*0.30)
    print("The Discounted price of a is",format(a))
    
```

```python
# 3. Using For loop Find the Factorial of a number

x = int(input("Enter a number: "))
fact = 1
if x<0:
    print("Negative numbers do not have a factorial. Please Enter a positive number.")
elif x==0:
    print("The factorial of the given number 0 is 1")
else:
    for i in range(1,x+1):
        fact=fact*i
    print("The factorial of the given number {x} is: "+str(fact))
```

```python
# 4. Fibonacci series (using for loop)

n = int(input("Till how many terms do you want to calculate the fibonacci series? "))

n1, n2 = 0, 1
if n<=0:
    print("Enter a positive number: ")
elif n==1:
    print(f"The Fibonacci series up to {n} is {n1}")
else:
    print("The Fibonacci series is: ")
    for i in range(n+1):
        print(n1, end=' ')
        n0=n1+n2
        n1=n2
        n2=n0
```

```python
# 5. Sum of first n numbers

n= int(input("Enter the value of n: "))
nos=n
if n<0:
    print("Enter a positive number: ")
else:
    sum=0
    while(n>0):
        sum +=n
        n-=1
    print(f"The sum of first {nos}"+" numbers is",sum)
```

```python
# 6. Reverse a number using while

n = int(input("Enter a number to be reversed: "))
rev_n = 0
while n!=0:
    i=n%10
    rev_n = rev_n*10+i
    n//=10
    
print("The reversed number is : ",rev_n)
```

```python
# 7. To Check if a number is an armstrong number or not

n = int(input("Enter a number: "))
sum=0
temp=n
while temp>0:
    i=temp%10
    sum+=i**3
    temp//=10
if n==sum:
    print(f"{n} is an Armstrong number")
else:
    print(f"{n} is not an Armstrong number")
```

```python
#8.Python program to print the following pattern
#*
#**
#***
#****
#*****
def pypart(n):
    for i in range(0, n):
        for j in range(0, i+1):
            print("* ",end="")
        print("\r")
n=5
pypart(5)
```
