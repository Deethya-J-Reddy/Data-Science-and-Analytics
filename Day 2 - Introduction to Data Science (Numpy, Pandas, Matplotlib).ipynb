# DATA SCIENCE

Data -Collection of information


 Quantittaive data:
        
      Data can be measurable
      Classification:
          Discrete or categorical or binary data
             Eg: Email Spam, Medical Diagnosis
          Regression - Continous Value
             Eg: Stock Price, Weather Prediction
Qualitative data:


     Data can't be measurable
          Nominal - eg: Gender
          Ordinal - eq: Company IDs of employees

 Statistics
 
 ->Measure of Central Tendency
 
     Mean-Total number of values/n
     Median-Middle
     Mode-Frequency of data

 ->Measure of Spread
 
     Standard Deviation
     Variance
     Range

 Sampling Process
    
    
     From the dataset we should collect some sample is called Sampling Process.
        
        
     Types:
          Random Sampling
          Stratified Sampling
          Simple Sampling Process

 Data Science Life Cycle

     1.Business Requirements
        -Requirement for the data
     2.Data Acqusition
        -Collect the data and understand
     3.Data Preprocessing or Cleaning or Wrangling
        -Cleaning the data
        -Corrupted Data
        -Missing values
        -Remove unwanted data
     4.Data Exploration
        -Explore the data and understand the hidden pattern of information
     5.Modelling
        -Analyze the Machine Learning Model
        -Apply the algorithm
     6.Data Deployment
        -Predict the final outcome for the requirements or Solution

 Steps for Data Science

     1.Collect the data-Kaggle
     2.Analyze the data-Pandas,Numpy
     3.Data Preprocessing
     4.Data visualization
     5.Apply the ML Model
     6.Predict the solution
